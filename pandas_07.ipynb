```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset('titanic')
```

```python
df_copy = df.copy()
```

```python
## groupby()
```

```python
# 성별 데이터 평균값
df_copy.groupby('sex').mean(numeric_only = True)
```

```python
# 성별 클래스등급별 데이터 평균
df_copy.groupby(['sex', 'pclass']).mean(numeric_only=True)
```

```python
# 성별 클래스 등급별 생존율
df_copy.groupby(['sex', 'pclass'])[['survived']].mean(numeric_only=True)
```

```python
## reset_index()
```

```python
df_temp = df_copy.groupby(['sex', 'pclass'])[['survived', 'age']].mean()
df_temp
```

```python
df_temp.reset_index()
```

```python
## pibot_table()
```

```python
df_copy.pivot_table(index='who', values = 'survived')
```

```python
df_copy.pivot_table(index=['who', 'pclass'], values = ['survived', 'age'])
```
