```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')
```

```python
## copy()
```

```python
df_copy = df.copy()
```

```python
## 결측치
1. 결측 데이터 확인
2. 결측치가 아닌 데이터 확인
3-1. 결측치 채우기
3-2. 결측 데이터 제거하기
```

```python
df_copy.isnull().sum() # 각 컬럼들의 결측치 개수 합산
# df_copy.isna().sum()
```

```python
df_copy.notnull().sum() # 각 컬럼들의 결측치가 아닌 값의 개수 합산
df_copy.notna().sum()
```

```python
## 결측 데이터 필터링
```

```python
cond = df_copy['age'].isnull()
df_copy.loc[cond, 'age'] = 30
df_copy.tail()
```

```python
## fillna()
```

```python
df_copy = df.copy()
df_copy['age'] = df_copy['age'].fillna(100)
df_copy.tail()
```

```python
df_copy['deck'].value_counts() # 각 값마다의 개수
```

```python
df_copy['deck'].fillna('C')
```

```python
df_copy['deck'] = df_copy['deck'].cat.add_categories('Z') # 카테고리에 Z추가
df_copy['deck'].fillna('Z')
```

```python
df_copy = df.copy()
age_mean = df_copy['age'].mean()

df_copy['age'] = df_copy['age'].fillna(age_mean)
df_copy.head()
```

```python
## 나이 결측치 채우기
- 남자 승객의 결측치는 남자 승객의 평균
- 여자 승객의 결측치는 여자 승객의 평균
- 결측치를 모두 채우고 나서 모든 사람의 평균
```

```python
df_copy = df.copy()
```

```python
cond1 = df_copy['sex'] == 'male'
cond2 = df_copy['sex'] == 'female'

male_age = df_copy.loc[cond1, 'age'].mean()
female_age = df_copy.loc[cond2, 'age'].mean()

df_copy.loc[cond1, 'age'] = df_copy.loc[cond1, 'age'].fillna(male_age)
df_copy.loc[cond2, 'age'] = df_copy.loc[cond2, 'age'].fillna(female_age)

df_copy['age'].mean()
```

```python
## dropna() : 결측치 제거
```

```python
df_copy = df.copy()
```

```python
df_copy.dropna()
```
