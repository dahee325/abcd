```python
## merge() : 병합
```

```python
info = {
    '고객명': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']
}

money = {
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']
}
```

```python
df1 = pd.DataFrame(info)
df2 = pd.DataFrame(money)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2) # 디폴트는 inner(교집합)
pd.merge(df1, df2, how='outer') # 합집합
pd.merge(df1, df2, how='left') # Left에 있는 df1행에 df2 출력
pd.merge(df1, df2, how='right') # Right 에 있는 df2행에 df1 출력
```
